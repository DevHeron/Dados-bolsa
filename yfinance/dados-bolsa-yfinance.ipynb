```python
# Do yfinance consigo: downloads; 
# Do investpy consigo a lista das empresas;
import yfinance as yf
import investpy as inv
import pandas as pd
```

```python
# Obtendo lista de empresas
br = inv.stocks.get_stocks(country='brazil')
```

```python
br
```

```python
br['symbol']
```

```python
carteira = []
for a in br['symbol']:
   if len(a) <= 5:
       carteira.append(a+".SA")

ticker = yf.Tickers(carteira)
```

```python
carteira
```

```python
dt_download_yf = yf.download(carteira, start='2019-01-01', end='2023-05-01')
```

```python
download_yf_df = pd.DataFrame.from_dict(dt_download_yf, orient='columns')
display(download_yf_df)
dt_download_yf.to_excel('download_yf.xlsx')
```
