```python
# Obter history e financial data do yahooquery, com tickers obtivos pelo investpy.
# documentação yahooquery: https://yahooquery.dpguthrie.com/
# documentação investpy: https://pypi.org/project/investpy/
```

```python
import pandas as pd
import pandas_datareader
import investpy as inv
from yahooquery import Ticker
```

```python
br = inv.stocks.get_stocks(country='brazil')
br
```

```python
br[['full_name','symbol']]
br.to_excel('nomes_empresas.xlsx')
```

```python
carteira = []
for a in br['symbol']:
    if a[-3] != '.SA':
        carteira.append(a+'.SA')

carteira
```

```python
# gerando excel com dados das empresas utilizadas para pesquisa:
br['symbol_sa'] = carteira
br.to_excel('empresas-info.xlsx')
br
```

```python
# Obtendo lista de tickers para usar na pesquisa:
tickers = Ticker(carteira, asynchronous=True)
```

```python
# histórico de preços da carteira de empresas brasileiras
history_br_df = tickers.history(start='2020-01-01', end='2023-10-01', interval='3mo')
```

```python
display(history_br_df)
history_br_df.to_csv('history_br_df.csv') #salvar em csv e converter no excel devido ao timezone
```

```python
# Financial data
types = ['NetIncome','FreeCashFlow']
fin_data = tickers.get_financial_data(types, trailing=False)
fin_data
```

```python
# Salvar fin_data (get_financial_data) em excel
fin_data.to_excel('fin_data.xlsx')
```

```python
#outra versão de financial data (financial_data)
fin_data2 = tickers.financial_data
fin_data2_df = pd.DataFrame.from_dict(fin_data2, orient='columns').T
```

```python
display(fin_data2_df)
```

```python
# Salvar financial_data no excel 
fin_data2_df.to_excel('fin_data2.xlsx')
```

```python
# Dados sobre ESG
esg_scores = tickers.esg_scores
#print(esg_scores)
```

```python
#Visualizando retorno da api sobre dados ESG para os tickers informados
esg_scores_df = pd.DataFrame.from_dict(esg_scores, orient="index")
display(esg_scores_df)
esg_scores_df.to_excel('esg_scores.xlsx')
```

```python
# variáveis para guardar earning history (histórico de recebimentos)
#OBS: já vem em dataframe!
earning_history = tickers.earning_history
display(earning_history)
```

```python
# - Salvando earning_history em excel
earning_history.to_excel('earn_history_br.xlsx')
```
